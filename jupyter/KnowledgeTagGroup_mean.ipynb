```python
import os
import pandas as pd
```

```python
data_dir = '/opt/ml/input/data/train_dataset'
file_name = 'train_all_three_basic_stats_ug_filtered.csv'
test_file_name = 'test_all_three_basic_stats_ug_filtered.csv'
```

```python
csv_file_path_to_load = os.path.join(data_dir, file_name)
df = pd.read_csv(csv_file_path_to_load)

csv_file_path_to_load = os.path.join(data_dir, test_file_name)
test_df = pd.read_csv(csv_file_path_to_load)
```

```python
df
```

```python
test_df
```

```python
KnowledgeTagGroup_answer_mean = df.groupby(['KnowledgeTagGroup'])['answerCode'].agg(['mean']).to_dict()
KnowledgeTagGroup_time_mean = df.groupby(['KnowledgeTagGroup'])['elapsed'].agg(['mean']).to_dict()

df['KnowledgeTagGroup_answer_mean'] = df.KnowledgeTagGroup.map(KnowledgeTagGroup_answer_mean['mean'])
df['KnowledgeTagGroup_time_mean'] = df.KnowledgeTagGroup.map(KnowledgeTagGroup_time_mean['mean'])

test_df['KnowledgeTagGroup_answer_mean'] = test_df.KnowledgeTagGroup.map(KnowledgeTagGroup_answer_mean['mean'])
test_df['KnowledgeTagGroup_time_mean'] = test_df.KnowledgeTagGroup.map(KnowledgeTagGroup_time_mean['mean'])
```

```python
train_name = f"{file_name.split('.')[0]}_gy_new.csv"
test_name = f"{test_file_name.split('.')[0]}_gy_new.csv"

df.to_csv(os.path.join(data_dir, train_name), sep=',', index=False)
test_df.to_csv(os.path.join(data_dir, test_name), sep=',', index=False)
```
